```python
import numpy as np
from matplotlib import pyplot as plt
from skimage import data
from skimage import util
from skimage import io
```

```python
# import the camera man image using data from skimage library
camera_man_img = data.camera()
# display the image to verify
plt.imshow(camera_man_img)
```

```python
# display it in grayscale
plt.imshow(camera_man_img, cmap='gray')
```

```python
# add some noise to the image : specifically "salt and pepper" noise
salty_peppered_image_hehe = util.random_noise(camera_man_img, mode='s&p', amount=0.1)
# display the image after adding some noise
plt.imshow(salty_peppered_image_hehe, cmap='gray')
```

```python
# compare the original image and after adding "salt and pepper" noise to it
# we need to use subplot :D
# original image UwU
plt.subplot(1, 2, 1)
plt.imshow(camera_man_img, cmap='gray')
plt.title("Mr Good Looking Camera Man")
# noisy image '-'
plt.subplot(1, 2, 2)
plt.imshow(salty_peppered_image_hehe, cmap='gray')
plt.title("Mr Blurry Looking Camera Man")

plt.show()
```

```python
# now we'll add a gaussian noise to the image :o
hugo_gauss_noise = util.random_noise(camera_man_img, mode='gaussian')
# display the difference between both
plt.subplot(1, 2, 1)
plt.imshow(camera_man_img, cmap='gray')
plt.title("Hugo Boss")

plt.subplot(1, 2, 2)
plt.imshow(hugo_gauss_noise, cmap='gray')
plt.title("Hugo Gauss")

plt.show()
```
